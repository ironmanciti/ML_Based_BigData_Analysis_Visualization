# Non Linear Regression Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5.0, 5.0, 0.1)

y = 2*(x) + 3  # 기울기 2, 절편 3
y_noise = 2 * np.random.normal(size=x.size)  # noise

ydata = y + y_noise

plt.figure(figsize=(8,6))
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

- 비선형 회귀는 독립 변수 $x$와 종속 변수 $y$ 사이의 관계로 비선형 함수 모델링 데이터를 생성합니다. 

- 본질적으로 선형이 아닌 관계는 비선형이라고 할 수 있으며 일반적으로 $k$ degree 다항식 ($x$의 최대 거듭 제곱)으로 표현됩니다.

$$ \ y = a x^3 + b x^2 + c x + d \ $$

비선형 함수는 지수, 로그, 분수 등과 같은 요소를 가질 수 있습니다. 예: $$ y = \log(x)$$
    
또는 다음과 같이 더 복잡합니다.
$$ y = \log(a x^3 + b x^2 + c x + d)$$

```python
x = np.arange(-5.0, 5.0, 0.1)

y = x**3 + x**2 + x + 3  # degree=3
y_noise = 20 * np.random.normal(size=x.size)  #noise
ydata = y + y_noise

plt.plot(x, ydata, 'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

다른 유형의 비선형 함수는 다음과 같습니다.
### Quadratic
$$ Y = X^2 $$

```python
x = np.arange(-5.0, 5.0, 0.1)

y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise

plt.plot(x, ydata, 'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

### Exponential

base가 c 인 지수 함수는 $$Y = a + b e^X$$로 정의됩니다. 

여기서 b ≠ 0, c> 0, c ≠ 1, x는 임의의 실수입니다. base c는 상수이고 지수 x는 변수입니다.

```python
X = np.arange(-5.0, 5.0, 0.1)

Y= 2 + 3 * np.exp(X)

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```
