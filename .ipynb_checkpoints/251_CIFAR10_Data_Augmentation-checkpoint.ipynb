# 데이터 증강으로 CIFAR-10 성능 향상

- CIFAR-10 및 CIFAR-100은 머신러닝과 컴퓨터 비전 알고리즘에 사용되는 8000만 개의 작은 이미지 데이터 세트의 하위 세트로 분류. 

- CIFAR-10은 32 x 32 크기의 60000개의 이미지로 이루어져 있다. 10개의 클래스로 분류(airplane, bird, cat ...) 되며 각각의 클래스는 6000개의 이미지로 이루어져 있다.
- 5000개는 학습 데이터, 1000개는 테스트 데이터 (총, 50000개의 학습 데이터, 10000개의 테스트 데이터)로 구성

- keras.preprocessing.image.ImageDataGenerator 이용

- flow method : 데이터 및 레이블 array를 가져와서 augmented data 배치를 생성

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, \
                        Dense, Activation, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import LearningRateScheduler
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

X_train_scaled = X_train / 255.
X_test_scaled  = X_test / 255.

y_train_onehot = to_categorical(y_train, 10)
y_test_onehot  = to_categorical(y_test, 10)

X_train.shape, y_train_onehot.shape, X_test.shape, y_test_onehot.shape
```

### ImageDataGenerator 메서드
- fit : 샘플 데이터에 데이터 생성기를 학습. 데이터 의존적인 변형에 관련된 데이터 통계를 계산. 
- flow : 데이터와 라벨 배열을 받아 증강된 데이터의 배치를 생성. (x, y) 튜플을 만들어내는 Iterator 반환

```python
# generator 생성
datagen = ImageDataGenerator(
            rotation_range=13,
            width_shift_range=0.1,
            height_shift_range=0.1,
            horizontal_flip=True
)

datagen.fit(X_train_scaled)
```

```python
cifar10_classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
from matplotlib import pyplot as plt

for X_batch, y_batch in datagen.flow(X_train_scaled, y_train_onehot, batch_size=9):
    # Show 9 images
    plt.figure(figsize=(6, 6))
    for i in range(9):
        plt.subplot(3, 3, i+1)
        plt.imshow((X_batch[i]*255).astype(np.uint8))
        plt.xticks([])
        plt.yticks([])
        plt.title(cifar10_classes[y_batch[i].argmax()])
    break
```

```python
img = plt.imread('datasets/banana.jpeg')
plt.imshow(img)

img = np.expand_dims(img, axis=0)
print(img.shape)
```

```python
# generate batches
aug_iter = datagen.flow(img)

# Get 10 samples of augmented images
aug_images = [next(aug_iter)[0].astype(np.uint8) for _ in range(10)]

plt.figure(figsize=(12, 4))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(aug_images[i])
    plt.xticks([])
    plt.yticks([])
```

```python
def lr_schedule(epoch):
    lr = 0.001
    if epoch > 75:
        lr = 0.0005
    if epoch > 100:
        lr = 0.0003
    return lr
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss',  patience=5)
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), padding='same', 
                 input_shape=(32, 32, 3), activation='relu'))

model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
 
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Conv2D(64, (3,3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.3))
 
model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())

model.add(Conv2D(128, (3,3 ), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))
 
model.add(Flatten())
model.add(Dense(10, activation='softmax'))

opt = tf.keras.optimizers.RMSprop(learning_rate=0.001,decay=1e-6)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(datagen.flow(X_train_scaled, y_train_onehot, batch_size=32), 
                        steps_per_epoch=len(X_train) // 32,
                        epochs=50, verbose=1,
                        validation_data=(X_test_scaled, y_test_onehot), 
                        callbacks=[LearningRateScheduler(lr_schedule), early_stop])
```

```python
model.evaluate(X_test_scaled, y_test_onehot, verbose=0)
```

```python
y_pred = model.predict(X_test_scaled).argmax(axis=1)
y_pred
```
